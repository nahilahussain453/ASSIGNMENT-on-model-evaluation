```python
age_median=data['Age'].median()
data['Age']=data['Age'].fillna(age_median)
```

```python
data.isna().sum()
```

```python
#Dropping
data=data.drop(['Name','Cabin','Ticket'],axis=1)
```

```python
embar_mode=data['Embarked'].mode()[0]
data['Embarked']=data['Embarked'].fillna(embar_mode)
data.isna().sum()
```

```python
data['Family']=data['SibSp']+data['Parch']
```

```python
data=data.drop(['SibSp','Parch'],axis=1)
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.iloc[:,2]=le.fit_transform(data.iloc[:,2].values)
data.iloc[:,5]=le.fit_transform(data.iloc[:,5].values)
```

```python
data.head()
```

```python
#Splitting data
x=data.iloc[:,1:]
y=data.iloc[:,0]
```

```python
#traintest split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#Modeling
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,classification_report
```

```python
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
acc=[]
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    acc.append(accuracy_score(y_test,y_pred))
    print("For k =",i,":",accuracy_score(y_test,y_pred))
```

```python
#accuracy score is 0.82
```

```python
#K fold cross validation
#importing k fold cross validation
from sklearn.model_selection import KFold
```

```python
kfold_validator=KFold(10)
```

```python
for train_index,test_index in kfold_validator.split(x,y):
    print("training Index:",train_index)
    print("test index",test_index)
```

```python
cv_result=cross_val_score(knn,x,y,cv=kfold_validator)
cv_result
```

```python
np.mean(cv_result)
```

```python
from sklearn.model_selection import StratifiedKFold
skfold_validator=StratifiedKFold(n_splits=10)
for train_index,test_index in skfold_validator.split(x,y):
    print("training Index:",train_index)
    print("test index",test_index)
```

```python
from sklearn.model_selection import cross_val_score
skcv_result=cross_val_score(knn,x,y,cv=skfold_validator)
np.mean(skcv_result)
```

```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
svm_lr=LinearSVC()
svm_lr.fit(x_train,y_train)
y_pred_svm_lr=svm_lr.predict(x_test)
accuracy_score(y_test,y_pred_svm_lr)
```

```python
svm_rbf=SVC(kernel='rbf')
svm_rbf.fit(x_train,y_train)
y_pred_svm2=svm_rbf.predict(x_test)
accuracy_score(y_test,y_pred_svm2)
```
